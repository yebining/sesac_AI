```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.array([2,4,6,8])
y=np.array([81,93,91,97])
```

```python
plt.scatter(x,y)
plt.plot(x,y,'r')
plt.show
```

```python
w=0
b=0

#학습률 임의로 0.03
lr = 0.03

epochs=2001
```

```python
#경사하강법
n=len(x)

#경사하강법 시작

for i in range(epochs):
    y_pred=w*x+b
    error = y-y_pred

    w_diff=(2/n)*sum(-x * (error)) #오차함수를 w로 편미분한 값
    b_diff=(2/n)*sum(-(error)) #오차함수를 b로 편미분한 값

    w=w-lr*w_diff #학습률 곱해서 기존의 w를 업데이트트
    b=b-lr*b_diff #학습률 곱해서 기존의 b를 업데이트트

    if i % 100 ==0:
        print('epochs=%.f, 기울기=%.4f, 절편=%.4f'%(i, w, b))
```

```python
y_pred=w*x+b

plt.scatter(x,y)
plt.plot(x,y_pred,'r')
plt.show
```

```python
x=np.array([[2,0], [4,4], [6,2], [8,3]])
y=np.array([81,93,91,97]) #y 실제받은성적
```

```python
model=Sequential()

model.add(Dense(1, input_dim=2, activation='linear'))

model.compile(optimizer='sgd', loss='mse')

#오차를 최소화하는 과정을 2000번 반복
model.fit(x,y, epochs=2001)

```

```python
#모델 테스트 
hour =7
private_class=4
prediction = model.predict([[hour, private_class]])

print('%.f 시간을 공부하고 %.f 시간의 과외를 받은 경우, 예상 점수는 %.02f 점입니다.' %(hour,  private_class, prediction))
```
